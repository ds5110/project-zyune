```python
import geopandas as gpd
Cumberland=gpd.read_file('Cumberland.geojson')
Cumberland=Cumberland[Cumberland.B19113_001E!=-666666666]
Cumberland
```

```python
aaa=Cumberland.explore(
     column="B19113_001E", # make choropleth based on "BoroName" column
     tooltip="B19113_001E", # show "BoroName" value in tooltip (on hover)
     popup=True, # show all values in popup (on click)
#      tiles="median income", # use "CartoDB positron" tiles
     cmap="Blues", # use "Set1" matplotlib colormap
     style_kwds=dict(color="grey"), # use black outline
    scheme='Percentiles'
)
aaa.save('family_median_income_Cumberland.geojson')
```
